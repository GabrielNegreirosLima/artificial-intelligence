A função desejada é `x²`, plotada abaixo.

```python
import numpy as np
import matplotlib.pyplot as plt

# 100 pontos igualmente espaçados entre -4 e 4
x = np.linspace(-4, 4, num=100)

# saída desejada
y_d = x**2

plt.plot(x,y_d, 'r', label='Função desejada: y=x²')
plt.legend()
plt.show()
```

Primeiramente precisamos iniciar com valores aleatórios todos os parâmetros:

```python
import random

def random_point():
    return random.random()*4 - 2 # returns between -2.0 and 2.0
```

Definimos as funções dos slides para calcular os parâmetros: 

```python
import math

# Dos slides
def y(p, x, q):
    return p*x + q

def line_function_1(x):
    return 1/2-(x/4)

def line_function_2(x):
    return 1/2+(x/4)

def update_p_1(p_1, alpha, error, x, w_1, w_2):
    return p_1 - alpha*error*x*(w_1/(w_1+w_2))

def update_p_2(p_2, alpha, error, x, w_1, w_2):
    return p_2 - alpha*error*x*(w_2/(w_1+w_2))

def update_q_1(q_1, alpha, error,w_1, w_2):
    return q_1 - alpha*error*(w_1/(w_1+w_2))

def update_q_2(q_2, alpha, error, w_1, w_2):
    return q_2 - alpha*error*(w_2/(w_1+w_2))

def calc_y(x_it, p_1, p_2, q_1, q_2):
    w_1 = line_function_1(x_it)
    w_2 = line_function_2(x_it)
    
    y_1 = p_1*x_it + q_1
    y_2 = p_2*x_it + q_2
    
    return ((w_1*y_1) + (w_2*y_2))/(w_1+w_2)

#f(-3, 0.7, 3, 0.7, 1.2, (1/2)+(1.2)/4, (1/2)-(1.2)/4)
calc_y(1.2, -3, 3, 0.7, 0.7)
```

Definimos um vetor com 1000 pontos aleatórios: 

```python
error = 1
max_it = 10000
it = 0
print(f"\n\n\nIT CARALHO: {it}")

#p_1 = random_point()
#q_1 = random_point()
#p_2 = random_point()
#q_2 = random_point()
p_1, q_1 = -3, 0.7
p_2,q_2= 3, 0.7
#p_1, q_1 = -1, 1
#p_2, q_2 = 1, 1
#print(p_1, p_2, q_1, q_2)

alpha = 0.1

x_axis = []
for _ in x:
    x_axis.append(random_point())

while (abs(error) > 0.001 or it==max_it):
    pos = random.randint(0, len(x)-1)
    x_it = x_axis[pos]
    w_1 = line_function_1(x_it)
    w_2 = line_function_2(x_it)
    
    y_f = calc_y(x_it, p_1, p_2, q_1, q_2)
    
    error = y_f - x_it**2
    print(f"error: {error}")
    y_vec = []
    for i in x:
        y_vec.append(calc_y(i, p_1, p_2, q_1, q_2))

    print(f"it:{it}\nx:{x_it}\np1:{p_1}\nq1:{q_1}\np2:{p_2}\nq2:{q_2}\nw1:{w_1}\nw2:{w_2}\nerro:{error}")
    plt.scatter(x, y_vec,marker=r'o',color='b', label="Aproximacao")
    plt.plot(x,y_d, 'r', label='Função desejada: y=x²')
    plt.legend()
    plt.show()
    
    p_1 = update_p_1(p_1, alpha, error, x_it, w_1, w_2)
    p_2 = update_p_2(p_1, alpha, error, x_it, w_1, w_2)
    
    q_1 = update_q_1(p_1, alpha, error, w_1, w_2)
    q_2 = update_q_2(p_1, alpha, error, w_1, w_2)
    
```
