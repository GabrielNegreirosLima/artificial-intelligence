A função desejada é `x²`, plotada abaixo.

```python
import numpy as np
import matplotlib.pyplot as plt

def get_points():
    points = np.random.uniform(low=-2, high=2, size=(100,))
    points.sort()
    
    return points

# saída desejada
points = get_points()
y_d = points**2


def plot_graph(points, y, label):
    
    plt.plot(points,y, 'r', label=label)
    plt.legend()
    plt.show()
    
plot_graph(points, y_d, 'Função desejada: y=x²')
```
