A função desejada é `x²`, plotada abaixo.

```python
import numpy as np
import matplotlib.pyplot as plt

def get_points():
    points = np.random.uniform(low=-2, high=2, size=(100,))
    points.sort()
    
    return points

# saída desejada
points = get_points()
y_d = points**2


def plot_graph(points, y, label, color):
    plt.plot(points,y, color=color, label=label)
    
def plot_combined(points, y_d, y_vec):
    plot_graph(points, y_d, 'Função desejada: y=x²', 'r')
    plot_graph(points, y_vec, 'Aproximação', 'b')
    plt.legend()
    plt.show()
    
plot_graph(points, y_d, 'Função desejada: y=x²', 'r')
```

```python
import math

def w(x, mean, sigma):
    return math.exp(-1/2 *((x-mean)/sigma)**2)

def y(p, x, q):
    return p*x + q

def final_y(x, p_1, p_2, q_1, q_2, mean_1, mean_2, sigma_1, sigma_2):
    w_1 = w(x, mean_1, sigma_1)
    w_2 = w(x, mean_2, sigma_2)
    y_1 = y(p_1, x, q_1)
    y_2 = y(p_2, x, q_2)
    
    return (w_1*y_1 + w_2*y_2)/(w_1 + w_2)

def create_y_vector(points, p_1, p_2, q_1, q_2, mean_1, mean_2, sigma_1, sigma_2):
    y = []
    
    for x in points:
        y.append(final_y(x, p_1, p_2, q_1, q_2, mean_1, mean_2, sigma_1, sigma_2))
    
    return y
```

```python
import random

alpha = 0.1
p_1 = np.random.uniform(low=-2, high=2)
p_2 = np.random.uniform(low=-2, high=2)
q_1 = np.random.uniform(low=-2, high=2)
q_2 = np.random.uniform(low=-2, high=2)
mean_1 = np.random.uniform(low=-2, high=2)
mean_2 = np.random.uniform(low=-2, high=2)
sigma_1 = np.random.uniform(low=-2, high=2)
sigma_2 = np.random.uniform(low=-2, high=2)

# A funcao inicial obtida é essa
y_vec = create_y_vector(points, p_1, p_2, q_1, q_2, mean_1, mean_2, sigma_1, sigma_2)
#plot_graph(points, y_vec, "aprox")
plot_combined(points, y_d, y_vec)
```

```python
error = 10
max_it = 500
it = 0 

y_vec = [] 
while abs(error) > 0.0001 and it < max_it:
    # 1. selecionar um ponto aleatório
    pos = np.random.randint(len(points) - 1)
    x = points[pos]
    
    y_it = final_y(x, p_1, p_2, q_1, q_2, mean_1, mean_2, sigma_1, sigma_2)
    error = y_it - y_d[pos]
    
    w_1 = w(x, mean_1, sigma_1)
    w_2 = w(x, mean_2, sigma_2)
    y_1 = y(p_1, x, q_1)
    y_2 = y(p_1, x, q_1)

    
    p_1 = p_1 - alpha*error*w_1/(w_1+w_2)*x
    p_2 = p_2 - alpha*error*w_2/(w_1+w_2)*x
    
    q_1 = q_1 - alpha*error*w_1/(w_1+w_2)
    q_2 = q_2 - alpha*error*w_2/(w_1+w_2)
    
    mean_1 = mean_1 - alpha*error*w_2*(y_1 - y_2)/((w_1+w_2)**2)*w_1*(x - mean_1)/(sigma_1**2)
    mean_2 = mean_2 - alpha*error*w_1*(y_2 - y_1)/((w_1+w_2)**2)*w_2*(x - mean_2)/(sigma_2**2)
    
    sigma_1 = sigma_1 - alpha*error*w_2*(y_1 - y_2)/((w_1+w_2)**2)*w_1*((x - mean_1)**2)/(sigma_1**3)
    sigma_2 = sigma_2 - alpha*error*w_1*(y_2 - y_1)/((w_1+w_2)**2)*w_2*((x - mean_2)**2)/(sigma_2**3)
    
    y_vec = create_y_vector(points, p_1, p_2, q_1, q_2, mean_1, mean_2, sigma_1, sigma_2)
    plot_combined(points, y_d, y_vec)

    print(f"error: {error} | it: {it}")
    it = it + 1
    
```
