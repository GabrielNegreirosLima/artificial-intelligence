Import do dataset da planta iris:

```python
from sklearn import datasets
import numpy as np
import random 

data = datasets.load_iris()
data
```

**Fun√ß√µes auxiliares para implementa√ß√£o do Perceptron:**

```python
def input_eval(W, x): #Avalia um elemento da amostra com base no conjunto de pesos
    
    _, LEN_WEIGHTS = W.shape
    BIAS_POS = LEN_WEIGHTS - 1 

    Weights = []
    Weights.append(W[0][0:len(W[0])-1])
    Weights.append(W[1][0:len(W[1])-1])
    Weights.append(W[2][0:len(W[2])-1])
    bias = []
    bias.append(W[0][BIAS_POS])
    bias.append(W[1][BIAS_POS])
    bias.append(W[2][BIAS_POS])

    result = np.zeros((len(W), 1))
  
    for i in range(0, len(W)):
        result[i][0] = np.dot(Weights[i], x) + bias[i]
    
    return result
```

```python
def update_weights(W, alpha, e, x): #Atualiza os pesos dos neur√¥nios
    
  # Weights update
    W[:, 0:4] += alpha * np.dot(e, x.reshape(1, -1))
  
  # Bias update
    W[:, 4] = W[:, 4] + np.squeeze(np.asarray((alpha * e)))
    return W
```

```python
def calc_error(u, expected): #Avalia o erro de classifica√ß√£o
    
    error = np.zeros((len(u), 1))
    for i in range(0, len(u)):
        if expected == i:
            error[i][0] = 1 - u[i][0]
        else:
            error[i][0] = 0 - u[i][0]
    return error
```

**Fun√ß√£o de ativa√ß√£o degrau:**
    
Dado o pertencimento avaliado utiliza a fun√ß√£o degrau para determinar a ativa√ß√£o do neur√¥nio.

```python
def step_activation(u): 
    
    return np.where(u >= 0, 1, 0)
```

**Fun√ß√£o Sigmoide:**
    
Dado o pertencimento avaliado utiliza a fun√ß√£o sigmoidal (tamb√©m conhecida como log√≠stica) para determinar a ativa√ß√£o do neur√¥nio.  

```python
def sigmoid_activation(t): 
    
    sigmoid_applied = 1/(1 + np.exp(-t))
    non_actived = np.zeros_like(sigmoid_applied)
    non_actived[sigmoid_applied.argmax(0), 0] = 1
    actived = non_actived
    return actived
```

**Implementa√ß√£o do Perceptron:**

```python
def Perceptron(max_it, alpha, X, D, activation_func):
    
    W = np.zeros((3,5))
    t = 1
    E = 1
    e = []
    Eacum = []
    
    while t < max_it and E > 0:
        E = 0
        poss = list(range(0, len(D)))
        
        for k in range(0, len(D)):
        
            j = random.randint(0, len(poss)-1)
            i = poss[j]
            poss.pop(j)

            y = None
        
            if(activation_func == 'sigmoid'):
                y = sigmoid_activation(input_eval(W, X[i]))
                print(y)
            elif(activation_func == 'step'):
                y = step_activation(input_eval(W, X[i]))   
                print(input_eval(W, X[i]))

            e.append(calc_error(y, D[i]))
            W = update_weights(W, alpha, e[len(e)-1], X[i])
            E += np.sum(np.square(e[len(e)-1]))
        
        t += 1
        Eacum.append(E)    
    return W
```

## Avalia√ß√£o

```python
from sklearn.model_selection import StratifiedKFold

skf = StratifiedKFold(shuffle=True, random_state=60)
X, y = data.data, data.target
```

```python
def model_evaluation(W, X, y, activation_func):
    correct = 0

    for i in range(0, len(y)):
        result = None
        if(activation_func == 'sigmoid'):
            result = sigmoid_activation(input_eval(W, X[i]))
        elif(activation_func == 'step'):
            result = step_activation(input_eval(W, X[i]))

        label = y[i]
        if result[label] == 1 and np.sum(result) == 1:
            correct += 1
    print(correct*100/len(y))
```

**Fun√ß√£o de ativa√ß√£o degrau:**

1000 Itera√ß√µes e Taxa de aprendizado = 0.1

```python
for train_index, test_index in skf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
  
    W = Perceptron(1000, 0.1, X_train, y_train, 'step')
    model_evaluation(W, X_test, y_test, 'step')
```

**Fun√ß√£o de ativa√ß√£o sigmoide:**

1000 Itera√ß√µes e Taxa de aprendizado = 0.1

```python
for train_index, test_index in skf.split(X, y):  
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
  
    W = Perceptron(1000, 0.1, X_train, y_train, 'sigmoid')
    model_evaluation(W, X_test, y_test, 'sigmoid')
```

## Comparativo dos resultados

A acur√°cia dos modelos gerados utilizando a fun√ß√£o sigmoide foi maior, com todos os resultados superiores a 90%, enquanto a fun√ß√£o step n√£o alcan√ßou mais de 80% de precis√£o.

Isso se d√° pois a fun√ß√£o log√≠stica (da fun√ß√£o sigmoide) avalia todo o espa√ßo amostral por ser uma fun√ß√£o cont√≠nua, enquanto a fun√ß√£o degrau (que n√£o √© cont√≠nua), √© mais brusca para problemas de classifica√ß√£o que utilizam fun√ß√µes de erro para determinar o passo de evolu√ß√£o.
