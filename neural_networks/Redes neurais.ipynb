## Perceptron com função de ativação degrau

```python
import numpy as np

iris_types = 3

# o bias estará dentro do W, por isso o tamanho é 5
# sendo o último elemento o bias
w_length = 5


def network_output(W, x):
    
    pos_bias = w_length - 1
    new_W = []
    bias = []
    W_size = len(W)
    network_output = np.zeros((W_size, 1))
    
    for w_it in range(0, iris_types):
        new_W.append(W[w_it][0:len(W[w_it])-1])
        bias.append(W[w_it][pos_bias])
  
    for i in range(0, W_size):
        network_output[i][0] = np.dot(new_W[i], x) + bias[i]

    return network_output


def perceptron_degrau(max_it=10, learning_rate=0.1, X=[], d=[]): 
    
    # o bias estará dentro do W, por isso o tamanho é 5
    W = np.zeros((iris_types,w_length))
    t = 1
    E = 1
    errors = []
    # Variavel n faz nada aparentemente: 
    # epoch_errors = []
    
    if len(X)==0 or len(d)==0:
        print("Entrada vazia', confira X e d")
        return []
    
    while t < max_it and E > 0:
        E = 0
        d_size = range(0, len(d))
        
        # O iterador antes era k
        for k in d_size:
        
            #j = random.randint(0, len(poss)-1)
            #i = poss[j]
            #poss.pop(j)
            
            # network output  
            y = None
            f = network_output(W, X[k])
            y = np.where(f>=0, 1, 0) 
            y_size = len(y)
            
            # error
            error_it = np.zeros((y_size, 1))
            for j in range(0, y_size):
                if d[j] == j:
                    error_it[j][0] = 1 - y[j][0]
                else:
                    error_it[j][0] = 0 - y[j][0]                   
            errors.append(error_it)
            last_error = errors[len(errors)-1]
                       
            # update weights and bias
            W[:, 0:4] += learning_rate*np.dot(last_error, X[k].reshape(1, -1))
            W[:, 4] = W[:, 4] + np.squeeze(np.asarray((learning_rate * last_error)))
            
            # accumulate error
            E += np.sum(np.square(last_error))
            print(E)
        
        t += 1
    return W
```
